*In tutti i casi, plottiamo il tempo normalizzato per singola somma degli esponenziali dei due elementi (di a e di b), cioè dividiamo i tempi ottenuti per p_max*

**Qui sotto, grafico 1.0:<<<2048,1024>>>; per p_max=200, 2000, 10000; sovrapposto al grafico CPU per p_max=2000**

*Si vede che CPU è lineare e lenta. Nota: somme senza exp"*

```python
import matplotlib.pyplot as plt
import csv
import numpy as np


data_gpu1 = np.genfromtxt("./datagpu31.txt", delimiter = ',')
data_gpu2 = np.genfromtxt("./datagpu32.txt", delimiter = ',')
data_gpu3 = np.genfromtxt("./datagpu33.txt", delimiter = ',')
data_cpu0 = np.genfromtxt("./datacpu40.txt", delimiter = ',')
#plot data
plt.figure(figsize=(20,20))
plt.plot(data_gpu1[:,0], data_gpu1[:,1]/200, label='p=200')
plt.plot(data_gpu2[:,0], data_gpu2[:,1]/2000, label='p=2000')
plt.plot(data_gpu3[:,0], data_gpu3[:,1]/10000, label='p=10000')
plt.plot(data_cpu0[:,0], data_cpu0[:,1]/2000, label='p=2000, CPU')
plt.xlabel("array length")
plt.ylabel(r' #$\mu$s for sum')
plt.legend(prop={'size': 20})
plt.show()
```
