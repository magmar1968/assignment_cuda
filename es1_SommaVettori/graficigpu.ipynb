**Qui sotto, grafico 1: p_max=500;step=60;<<<1,256>>>; sovrapposto a p_max=500, step=150 e <<<4,1024>>>***

*Si vede che aumetando il numero di thread e blocchi, la velocit√† aumenta e "ci sono meno gradini"*

```python
import matplotlib.pyplot as plt
import csv
import numpy as np


data_gpu1 = np.genfromtxt("./datagpu1.txt", delimiter = ',')
data_gpu14 = np.genfromtxt("./datagpu14.txt", delimiter = ',')
#plot data
plt.figure(figsize=(20,20))
plt.plot(data_gpu1[:,0], data_gpu1[:,1]/500, label='<<<1,256>>>')
plt.plot(data_gpu14[:,0], data_gpu14[:,1]/500, label='<<<4,1024>>>')
plt.xlabel("array length")
plt.ylabel(r' #$\mu$s for sum')
plt.legend(prop={'size': 20})
plt.show()
```
