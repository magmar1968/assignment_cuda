**Generatore gaussiano, 10^6 numeri, media 0, std_deviation 20**

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
import statsmodels.api as sm
import pylab
from scipy import stats
```

```python
campioniGauss1 = np.genfromtxt("./campioniGauss1.txt", delimiter = 'None')
```

```python
campioniGauss1 = np.genfromtxt("./campioniGauss1.txt", delimiter = 'None')

#plot data
plt.figure(figsize=(10,5))
plt.hist(campioniGauss1, density = True, stacked = True, bins = 75)
plt.xlabel("Classi di numeri generati")
plt.ylabel('Frequenze relative')

plt.show()
```

```python
#calcolo media
media = np.mean(campioniGauss1)
media
```

```python
#test sulla media
stats.ttest_1samp(campioniGauss1, popmean = 0)
```

```python
#deviazione standard
std=np.sqrt(np.var(campioniGauss1))
std
```

***QQ-Plot***

```python
fig = sm.qqplot((campioniGauss1-media)/std, line='45')
plt.show()
```

**Generatore lineare congruente: 10^7 numeri da genUniform, intervallo (0,1)**

```python
campioniUniform1 = np.genfromtxt("./campioniUniform1.txt", delimiter = 'None')
```

```python
campioniUniform1 = np.genfromtxt("./campioniUniform1.txt", delimiter = 'None')

#plot data

plt.figure(figsize=(10,5))
plt.hist(campioniUniform1, density = True, stacked = True, bins = 50)
plt.xlabel("Classi di numeri generati")
plt.ylabel('Frequenze relative')
ax = plt.gca()
plt.show()
```

```python
#media
media = np.mean(campioniUniform1)
media
```

```python
#test sulla media
stats.ttest_1samp(campioniUniform1, popmean = 0.5)
```

```python
#deviazione standard
std=np.sqrt(np.var(campioniUniform1))
std
```

```python
#test: la distribuzione  normale?
stats.normaltest(campioniUniform1)
```

```python
#test: la distribuzione  uniforme? (Kolmogorov-Smirnov)
stats.kstest(campioniUniform1,stats.uniform(loc = 0, scale =1).cdf)
```

***QQ-Plot (per confronto con quanto fatto con gaussiani)***

*Ha una forma ad 's' come ci si aspetta per una distribuzione uniforme*

```python
with plt.rc_context():
    plt.rc("figure", figsize=(10,5))
    fig = sm.qqplot((campioniUniform1-media)/std, line='45')
plt.show()
```
