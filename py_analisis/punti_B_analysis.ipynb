```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker
```

```python
pathoutfile = "../pricer/data/outfile_exam/"
```

### Punto B3a

```python
df = pd.read_csv(pathoutfile + "outfile_B3a.csv")

m_values = df["m"].to_numpy()
exact_vals = df["exact_result"].to_numpy()
exact_err  = df["exact_error"].to_numpy()
approx_vals = df["approx_result"].to_numpy()
approx_err  = df["approx_erro"].to_numpy()
scale = 1.2

plt.plot(m_values,approx_vals,"-o",label = r"pay off approx")
plt.plot(m_values,exact_vals, "-o",label = r"pay off exact ")
plt.ylabel(r"Pay off", size = 12*scale)
plt.xlabel(r"m",       size = 12*scale)
plt.xticks(size = 10*scale)
plt.yticks(size = 10*scale)
plt.tick_params(axis="both", direction='in', length=6*scale,width=1*scale)
plt.legend( prop={'size': 10*scale})

plt.show()
```

### Punto B3b

```python
data_k100 = np.genfromtxt(pathoutfile + "outfile_B3b_K1.txt"  , delimiter = ',', comments='#')
data_k050 = np.genfromtxt(pathoutfile + "outfile_B3b_K05.txt"  , delimiter = ',', comments='#')
data_k075 = np.genfromtxt(pathoutfile + "outfile_B3b_K075.txt", delimiter = ',', comments='#')

B_values = data_k050[:,0]
scale = 1.2

plt.plot(B_values,data_k100[:,1],"-o",label = "K = 1"   )
plt.plot(B_values,data_k075[:,1],"-o",label = "K = 0.75")
plt.plot(B_values,data_k050[:,1],"-o",label = "K = 0.50")

plt.ylabel(r"Pay off",size=12*scale)
plt.xlabel(r"B",size=12*scale)
plt.xticks(size = 10*scale)
plt.yticks(size = 10*scale)
plt.tick_params(axis="both", direction='in', length=6*scale,width=1*scale)
plt.legend( prop={'size': 10*scale})
plt.show()

plt.plot(B_values,data_k100[:,2],"-o",label = "K = 1   ")
plt.plot(B_values,data_k075[:,2],"-o",label = "K = 0.75")
plt.plot(B_values,data_k050[:,2],"-o",label = "K = 0.50")

plt.ylabel(r"MC error",size=12*scale)
plt.xlabel(r"B",size=12*scale)
plt.xticks(size = 10*scale)
plt.yticks(size = 10*scale)
plt.tick_params(axis="both", direction='in', length=6*scale,width=1*scale)
plt.legend( prop={'size': 10*scale})
plt.show()

```

### Punto B4a

```python
df = pd.read_csv(pathoutfile + "outfile_B3a.csv")

m_values = df["m"].to_numpy()
exact_vals = df["exact_result"].to_numpy()
exact_err  = df["exact_error"].to_numpy()
appro_vals = df["approx_result"].to_numpy()
appro_err  = df["approx_erro"].to_numpy()
scale = 1.2

exact_err_percent = (exact_err * 100) /exact_vals
appro_err_percent = (appro_err * 100) /exact_vals

plt.plot(m_values,appro_err_percent,"-o",label = r"pay off approx")
plt.plot(m_values,exact_err_percent,"-o",label = r"pay off exact ")
plt.ylabel(r"$\epsilon_{MC}$ %", size = 12*scale)
plt.xlabel(r"m",       size = 12*scale)
plt.xticks(size = 10*scale)
plt.yticks(size = 10*scale)
plt.tick_params(axis="both", direction='in', length=6*scale,width=1*scale)
plt.legend( prop={'size': 10*scale})

plt.show()
```

### Punto B4b

```python
data = np.genfromtxt(pathoutfile + "outfile_B4b.txt"  , delimiter = ',', comments='#')
Nsim   = data[:,0]
MC_err = data[:,2]

#first part
plt.loglog(Nsim,MC_err , "-o" )
plt.ylabel(r"$\epsilon_{MC}$", size = 12*scale)
plt.xlabel(r"N",       size = 12*scale)
plt.xticks(size = 10*scale)
plt.yticks(size = 10*scale)
plt.tick_params(axis="both", direction='in', length=6*scale,width=1*scale)
plt.tick_params(which = "minor", axis="both",direction="in", length= 4*scale,width=0.5*scale)
plt.show()


#second part
NSIM = 15000

plt.plot(B_values,data_k100[:,2]*np.sqrt(NSIM),"-o",label = "K = 1   ")
plt.plot(B_values,data_k075[:,2]*np.sqrt(NSIM),"-o",label = "K = 0.75")
plt.plot(B_values,data_k050[:,2]*np.sqrt(NSIM),"-o",label = "K = 0.50")

plt.ylabel(r"H(B)",size=12*scale)
plt.xlabel(r"B",size=12*scale)
plt.xticks(size = 10*scale)
plt.yticks(size = 10*scale)
plt.ticklabel_format(style='sci', axis='y',scilimits=(0,0))
plt.tick_params(axis="both", direction='in', length=6*scale,width=1*scale)
plt.legend( prop={'size': 10*scale})
plt.show()
```

### Punto B5a

```python
data_m2   = np.genfromtxt(pathoutfile + "outfile_B5_m2.txt",  delimiter = ',', comments='#')
data_m5   = np.genfromtxt(pathoutfile + "outfile_B5_m5.txt",  delimiter = ',', comments='#')
data_m50  = np.genfromtxt(pathoutfile + "outfile_B5_m50.txt", delimiter = ',', comments='#')
data_m200 = np.genfromtxt(pathoutfile + "outfile_B5_m200.txt",delimiter = ',', comments='#')

scale = 2
fig,axs = plt.subplots(2, 2, sharex=True, sharey=False,figsize=(8*scale,6*scale))
axs[0,0].set_xticks(np.arange(0,601,120))

axs[0,0].plot(data_m2[:,0],data_m2[:,3],label = " m = 2")
axs[0,0].legend(prop={'size': 10*scale})
axs[0,0].grid()
axs[0,0].tick_params(axis='both', which='major', labelsize=6*scale)

axs[0,1].plot(data_m5[:,0],data_m5[:,3],label = " m = 5")
axs[0,1].legend(prop={'size': 10*scale})
axs[0,1].grid()
axs[0,1].tick_params(axis='both', which='major', labelsize=6*scale)

axs[1,0].plot(data_m50[:,0],data_m50[:,3],label = " m = 50")
axs[1,0].legend(prop={'size': 10*scale})
axs[1,0].grid()
axs[1,0].tick_params(axis='both', which='major', labelsize=6*scale)

axs[1,1].plot(data_m200[:,0],data_m200[:,3],label = " m = 200")
axs[1,1].legend(prop={'size': 10*scale})
axs[1,1].grid()
axs[1,1].tick_params(axis='both', which='major', labelsize=6*scale)


#add hided plot for labelling axes
fig.add_subplot(111, frameon=False)
plt.tick_params(labelcolor='none', which='both', top=False, bottom=False, left=False, right=False)
plt.xlabel(r"N BLOCKS", size = 10*scale)
plt.ylabel(r"$t_{GPU} [s]$", size = 10*scale)

plt.show()

```
