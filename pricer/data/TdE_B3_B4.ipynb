 **B3.b)  Prezzo in funzione del parametro B**

```python
import matplotlib.pyplot as plt
import csv
import numpy as np


data_b3bk1  = np.genfromtxt("./out_B3b_K1.txt", delimiter = ',', comments='#')
data_b3bk05 = np.genfromtxt("./out_B3b_K05.txt", delimiter = ',', comments='#')
data_b3bk075 = np.genfromtxt("./out_B3b_K075.txt", delimiter = ',', comments='#')

# plot data
plt.figure(figsize=(8, 6), dpi=100, facecolor=(1, 1, 1))

plt.plot(data_b3bk1[:,0], data_b3bk1[:,1] , marker='.', label = 'k = 1')
plt.plot(data_b3bk075[:,0], data_b3bk075[:,1] , marker='.', label = 'k = 0.75')
plt.plot(data_b3bk05[:,0], data_b3bk05[:,1] , marker='.', label = 'k = 0.5')



plt.xlabel("B", fontsize = 13)
plt.ylabel(r'Price', fontsize = 13)
plt.legend()
plt.title("Prezzo in funzione del parametro B", fontsize = 15)
plt.savefig('B3b.png')
plt.show()
```

**B4.b) Parte iniziale: errore in funzione di N**

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from scipy import stats


data_b4bN = np.genfromtxt("./out_B4.txt", delimiter = ',', comments='#')


# plot data
plt.figure(figsize=(8, 6), dpi=100, facecolor=(1, 1, 1))

plt.plot(np.log(data_b4bN[:,0]), np.log(data_b4bN[:,2]) , marker='o', label = 'MC error as a function of N')


plt.xlabel("log N", fontsize = 13)
plt.ylabel(r'log $\epsilon$    (absolute)', fontsize = 13)
plt.title("Errore montecarlo in funzione del parametro N", fontsize = 15)
plt.savefig('B4bN.png')
plt.show()
slope, intercept, r_value, p_value, std_err = stats.linregress(np.log(data_b4bN[:,0]),  np.log(data_b4bN[:,2]))
```

**B4.b) Parte finale: errore in funzione di B**

```python
import matplotlib.pyplot as plt
import csv
import numpy as np



# plot data
plt.figure(figsize=(8, 6), dpi=100, facecolor=(1, 1, 1))

plt.plot(data_b3bk1[:,0], data_b3bk1[:,2]*np.sqrt(15000) , marker='.', label = 'K=1')
plt.plot(data_b3bk075[:,0], data_b3bk075[:,2]*np.sqrt(15000) , marker='.', label = 'K=0.75')
plt.plot(data_b3bk05[:,0], data_b3bk05[:,2]*np.sqrt(15000) , marker='.', label = 'K=0.5')


plt.xlabel("B", fontsize = 13)
plt.ylabel(r'H(B)', fontsize = 13)
plt.title("Errore montecarlo in funzione del parametro B", fontsize = 15)
plt.legend()
plt.savefig('B4b.png')
plt.show()
```
