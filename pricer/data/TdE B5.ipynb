```python
import matplotlib.pyplot as plt
import csv
import numpy as np


data1 = np.genfromtxt("./out_B5_GPU_m200_long_range.txt", delimiter = ',', comments='#')
data2 = np.genfromtxt("./out_B5_GPU_m200_long_range_512TPB.txt", delimiter = ',', comments='#')

# plot data
plt.figure(figsize=(8, 6), dpi=100)

plt.plot(data1[:,0], data1[:,3] , marker='o', label = '32 TPB', color = 'blue')
plt.plot(data2[:,0], data2[:,3] , marker='o', label = '512 TPB', color = 'orange')

plt.xlabel("# blockss", fontsize = 13)
plt.ylabel(r'Time (s)', fontsize = 13)
plt.title("Computation time as # blocks increases", fontsize = 15)
plt.savefig('B5_m200.png')
plt.show()
```
