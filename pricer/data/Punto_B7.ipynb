```python
import matplotlib.pyplot as plt
import numpy as np
dataB7_m1 = np.genfromtxt('out_B7_m1.txt', delimiter = ',', comments = '#')
dataB7_m10 = np.genfromtxt('out_B7_m10.txt', delimiter = ',', comments = '#')
dataB7_m100 = np.genfromtxt('out_B7_m100.txt', delimiter = ',', comments = '#')

plt.figure(figsize=(8,5))
plt.title("Probability P(B)")
plt.plot(dataB7_m1[:,0], dataB7_m1[:,1], label = "m = 1", marker = '.')
plt.plot(dataB7_m10[:,0], dataB7_m10[:,1], label = "m = 10", marker = '.')
plt.plot(dataB7_m100[:,0], dataB7_m100[:,1], label = "m = 100", marker = '.')
plt.xlabel('B')
plt.ylabel('P(B)', rotation = 0)
plt.legend()

plt.savefig("B7_plot1.png")
```

```python
plt.figure(figsize=(8,5))
plt.title("Montecarlo error $\epsilon_{MC}(B)$")
plt.plot(dataB7_m1[:,0], dataB7_m1[:,2], label = "m = 1", marker = '.')
plt.plot(dataB7_m10[:,0], dataB7_m10[:,2], label = "m = 10", marker = '.')
plt.plot(dataB7_m100[:,0], dataB7_m100[:,2], label = "m = 100", marker = '.')
plt.xlabel('B')
plt.ylabel('$\epsilon_{MC}$', rotation = 0)
plt.legend()

plt.savefig("B7_plot2.png")
```

```python
plt.figure(figsize=(8,5))
plt.title("Errore montecarlo $\epsilon_{MC}(P)$")
plt.plot(dataB7_m1[:,1], dataB7_m1[:,2], label = "m = 1", marker = '.')
plt.plot(dataB7_m10[:,1], dataB7_m10[:,2], label = "m = 10", marker = '.')
plt.plot(dataB7_m100[:,1], dataB7_m100[:,2], label = "m = 100", marker = '.')
plt.xlabel('P')
plt.ylabel('$\epsilon_{MC}$', rotation = 0)
plt.legend()

plt.savefig("B7_plot3.png")
```
